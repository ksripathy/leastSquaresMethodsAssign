```python
import numpy as np
import matplotlib.pyplot as plt



n = 101
mean = 0
std_dev = 1

vec1 = np.random.normal(mean, std_dev, n)
vec2 = np.random.normal(mean, std_dev, n)

pdf_vec1 = 1/(std_dev * np.sqrt(2 * np.pi)) * np.exp( - (np.sort(vec1) - mean)**2 / (2 * std_dev**2))
pdf_vec2 = 1/(std_dev * np.sqrt(2 * np.pi)) * np.exp( - (np.sort(vec2) - mean)**2 / (2 * std_dev**2))

gridVec1, gridVec2 = np.meshgrid(vec1, vec2)
jointpdf = 1/(std_dev * np.sqrt(2 * np.pi)) * np.exp( - (gridVec1 - mean)**2 / (2 * std_dev**2)) * 1/(std_dev * np.sqrt(2 * np.pi)) * np.exp( - (gridVec2 - mean)**2 / (2 * std_dev**2))

fig1, ax1 = plt.subplots()
ax1.plot(vec1, pdf_vec1, 'o', label='vec1')
ax1.plot(vec2, pdf_vec2, 'o', label='vec2')

x1 = np.linspace(-3, 3, 101)
x2 = x1
gridX1, gridX2 = np.meshgrid(x1, x2)
jointpdf1 = 1/(std_dev * np.sqrt(2 * np.pi)) * np.exp( - (gridX1 - mean)**2 / (2 * std_dev**2)) * 1/(std_dev * np.sqrt(2 * np.pi)) * np.exp( - (gridX2 - mean)**2 / (2 * std_dev**2))

pdf1 = 1/(std_dev * np.sqrt(2 * np.pi)) * np.exp( - (x1 - mean)**2 / (2 * std_dev**2))

print(np.trapz(pdf1[:51], x1[:51]))
print(np.trapz(jointpdf1[50,:51], gridX1[50,:51]))




```
